```python
class dt(): 
    def __init__(self,model,data): 
        filename = '{0}_0.5p_{1}_latih_sampled30_'.format(model,data)
        df1 = pd.read_excel("%s1.xlsx"%filename)
        df2 = pd.read_excel("%s2.xlsx"%filename)
        df3 = pd.read_excel("%s3.xlsx"%filename)
        df4 = pd.read_excel("%s4.xlsx"%filename)
        df5 = pd.read_excel("%s5.xlsx"%filename)
        self.d1=df1
        self.d2=df2
        self.d3=df3
        self.d4=df4
        self.d5=df5
def rdmlp(data): 
    return dt('mlp',data) 
def rdrf(data): 
    return dt('rf',data) 
def rdrf2(data): 
    return dt('rf_2',data)
def rdrf3(data): 
    return dt('rf_3',data) 
def rdsvm(data): 
    return dt('svm',data) 
def rdsvm2(data): 
    return dt('svm_2',data) 
```

```python
def readdtmlp(data):
    filename = 'mlp_0.5p_%s_modelhasilsmote5'%data
    df = pd.read_excel("%s.xlsx"%filename)
    #df=df.drop(columns='Unnamed: 0')
    return df
def readdtrf(data):
    filename = 'rf_100n_1000m_0.5p_%s_modelhasilsmote2'%data
    df = pd.read_excel("%s.xlsx"%filename,header=None)
    #df=df.drop(columns='Unnamed: 0')
    return df
def readdtsvm(data):
    filename = 'svm_rbf_0.5p_%s_modelhasilsmote5'%data
    df = pd.read_excel("%s.xlsx"%filename)
    #df=df.drop(columns='Unnamed: 0')
    return df
```

```python
def mergehasil(data,model):
    if model=='mlp':
        dtf_mlp = rdmlp(data)
    elif model=='rf':
        dtf_mlp = rdrf(data)
    elif model=='rf_2':
        dtf_mlp = rdrf2(data)
    elif model=='rf_3':
        dtf_mlp = rdrf3(data)
    elif model=='svm':
        dtf_mlp = rdsvm(data)
    elif model=='svm_2':
        dtf_mlp = rdsvm2(data)
    def f(row):
        if row['score'] >= 0.5:
            val = 1
        else:
            val = 0
        return val
    column_names = ['com','prot','score','class']
    ndtf_mlp = pd.DataFrame(columns = column_names)
    ndtf_mlp['com']=dtf_mlp.d1[2]
    ndtf_mlp['prot']=dtf_mlp.d1[3]
    ndtf_mlp['score']=pd.concat([dtf_mlp.d1[4],dtf_mlp.d2[4],dtf_mlp.d3[4],dtf_mlp.d4[4],dtf_mlp.d5[4]],
                                axis=1).mean(axis=1)
    ndtf_mlp['class']= ndtf_mlp.apply(f, axis=1)
    ndtf_mlp.to_excel('{0}_{1}.xlsx'.format(data,model))
    return ndtf_mlp
```

```python
class dt2(): 
    def __init__(self,data): 
        df1 = pd.read_excel("%s_mlp.xlsx"%data)
        df2 = pd.read_excel("%s_rf.xlsx"%data)
        df3 = pd.read_excel("%s_svm_2.xlsx"%data)
        self.d1=df1
        self.d2=df2
        self.d3=df3
def rd(data): 
    return dt2(data) 
```

```python
def mergehasil2(data):
    dtf_mlp=rd(data)
    column_names = ['com','prot','score','class']
    ndtf_mlp = pd.DataFrame(columns = column_names)
    ndtf_mlp['com']=dtf_mlp.d1['com']
    ndtf_mlp['prot']=dtf_mlp.d1['prot']
    ndtf_mlp['score']=pd.concat([dtf_mlp.d1['score'],dtf_mlp.d2['score'],dtf_mlp.d3['score']],
                                axis=1).mean(axis=1)
    ndtf_mlp['class']= ndtf_mlp.apply(f, axis=1)
    ndtf_mlp['mlp'] = dtf_mlp.d1.apply(f,axis=1)
    ndtf_mlp['mlp_score']=dtf_mlp.d1['score']
    ndtf_mlp['rf'] = dtf_mlp.d2.apply(f,axis=1)
    ndtf_mlp['rf_score']=dtf_mlp.d2['score']
    ndtf_mlp['svm'] = dtf_mlp.d3.apply(f,axis=1)
    ndtf_mlp['svm_score']=dtf_mlp.d3['score']
    ndtf_mlp['count']=pd.concat([ndtf_mlp['mlp'],ndtf_mlp['rf'],ndtf_mlp['svm']],axis=1).sum(axis=1)
    ndtf_mlp.to_excel('{0}_hasilmerge.xlsx'.format(data))
    return ndtf_mlp
```
