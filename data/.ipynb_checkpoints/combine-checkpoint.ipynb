```python
def combine_data(X,Y,Z):
    compound = X
    protein = Y
    dt_prot = pd.read_csv(protein, delimiter=";")
    dt_comp = pd.read_excel(compound, header=None)
    prots = dt_prot.values
    comps = dt_comp.values

    cxp = concate_data(comps,prots)
    print(dt_prot.shape)
    print(dt_comp.shape)
    print(dt_comp.shape[1]+dt_prot.shape[1])
    np.savetxt(Z,cxp,fmt=["%s"]*(1283), delimiter=";")
    return cxp ### ["%d",] ["%.5f",]

def concate_data(A,B):
    row1,col1 = A.shape
    row2,col2 = B.shape

    out = np.zeros((row1,row2,col1+col2), dtype = B.dtype)
    out[:,:,:col1] = A[:,None,:]
    out[:,:,col1:] = B

        
    return out.reshape(row1*row2,-1)
```
