```python
from sklearn import model_selection
```

```python
from imblearn.over_sampling import RandomOverSampler
oversample = RandomOverSampler(sampling_strategy=0.9)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy=0.9)
```

```python
#X_postrain, X_postest, Y_postrain, Y_postest = model_selection.train_test_split(X_pos, Y_pos, test_size=0.7, random_state=seed)
#X_negtrain, X_negtest, Y_negtrain, Y_negtest = model_selection.train_test_split(X_neg, Y_neg, test_size=0.3, random_state=seed)
```

```python
def part_latih(seed):
    X_postrain, X_postest, Y_postrain, Y_postest = model_selection.train_test_split(X_neg, Y_neg, test_size=0.7, random_state=seed)

    X3 = X_postrain.append(X_pos)
    Y3 = Y_postrain.append(Y_pos)
    
    X_over, y_over = oversample.fit_resample(X3,Y3)
    X_over=pd.DataFrame(X_over)
    y_over=pd.DataFrame(y_over)
    
    X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X_over, y_over, test_size=0.3, random_state=seed)
    X_train=pd.DataFrame(X_train)
    Y_train=pd.DataFrame(Y_train)
    X_test=pd.DataFrame(X_test)
    Y_test=pd.DataFrame(Y_test)
    
    XXX_train=pd.concat([X_train,Y_train], axis=1)
    XXX_train.to_csv("latih_sampled30_"+str(seed)+".csv",index=False)
    
    XXX_test=pd.concat([X_test,Y_test], axis=1)
    XXX_test.to_csv("uji_sampled30_"+str(seed)+".csv",index=False)
    
    #return XXX
```

```python
for i in range(1,6):
    part_latih(i)
```
